```python
# rename colums
df.rename(columns = {'photoUrl':'Photo URL', 'LongName':'Long Name', 'playerUrl':'Player URL', 'foot':'Foot', '↓OVA':'OVA'}, inplace = True)
```

```python
df['Team & Contract'] = df['Team & Contract'].astype('str') #change type to str
```

```python
df['Team & Contract'].replace('\n', '', regex = True, inplace = True) # replace values '\n' with ''
```

```python
df['Team & Contract'][1] #display values rows 1 from colum 'Team & Contract'
```

```python
# We're going to separate 'Team & Contract' into columns 'Club' and 'Contract Time'
Club = []
Contract_time = []
for x in range(len(df['Team & Contract'])):
    Club.append(str(df['Team & Contract'][x][:-11]))
    
    ct = str(df['Team & Contract'][x][-11:])
    if ct.startswith('2') == True:
        Contract_time.append(ct)
    else:
        Contract_time.append('0')
df['Club'] = Club
df['Contract Time'] = Contract_time
```

```python
# Cleaning columns 'Hits' and change the type to int
hit = []
for x in range(len(df['Hits'])):
    a = str(df['Hits'][x])
    a = a.replace('\n','')
    a = a.replace('K','00')
    a = a.replace('.','')
    hit.append(a)
df['Hits'] = hit
df['Hits'].astype(int)
```

```python
# create function for changeging format columns 'Joined'
def month_join(x):
    if x=='Jan':
        return 1
    elif x=='Feb':
        return 2
    elif x=='Mar':
        return 3
    elif x=='Apr':
        return 4
    elif x=='May':
        return 5
    elif x=='Jun':
        return 6
    elif x=='Jul':
        return 7
    elif x=='Aug':
        return 8
    elif x=='Sep':
        return 9
    elif x=='Oct':
        return 10
    elif x=='Nov':
        return 11
    elif x=='Dec':
        return 12
```

```python
# cleaning and change format from column 'Joined'
Date = []
for x in range(len(df['Joined'])):
    a = df['Joined'][x]
    b = a.split(' ')
    month = str(month_join(b[0]))
    day = str(b[1].replace(',',''))
    year = str(b[2])
    if len(day)==1:
        day = ('0'+str(day))
    date_long = (str(month)+'/'+str(day)+'/'+str(year))
    date_con = pd.to_datetime(date_long)
    Date.append(date_con)
```

```python
df['Joined'] = Date
df['Joined']
```

```python
# Cleaning and change type from column height to int
height = []
for x in range(len(df['Height'])):
    a = str(df['Height'][x])
    a = a.replace('"','').split("'")
    a = int(a[0])*12 + int(a[-1])
    height.append(a)
df['Height'] = height
df['Height'].astype(int)
```

```python
# Cleaning and change type from column height to int
weight = []
for x in range(len(df['Weight'])):
    weight.append(int(df['Weight'][x].split('lbs')[0]))
df['Weight'] = weight
df['Weight']
```

```python
# Cleaning and change type from column 'Value' to int
value = []
for x in range(len(df['Value'])):
    v = (df['Value'][x])
    v = v.replace('€','').replace('K','00').replace('M','000000').replace('.','d')
    if 'd' in v:
        v = v.replace('d','')
        v = int(v)/10
    value.append(int(v))
df['Value'] = value
df['Value']
```

```python
# Cleaning and change type from column 'wage' to int
wage = []
for x in range(len(df['Wage'])):
    w = df['Wage'][x]
    w = w.replace('€','').replace('K','00')
    wage.append(int(w))
df['Wage'] = wage
df['Wage']
```

```python
# Cleaning and change type from column 'Release Clause' to int
recl = []
for x in range(len(df['Release Clause'])):
    rc = df['Release Clause'][x]
    rc = rc.replace('€','').replace('M','000000').replace('K','00').replace('.','d')
    if 'd' in rc:
        rc = rc.replace('d','')
        rc = int(rc)/10
    recl.append(int(rc))
df['Release Clause'] = recl
df['Release Clause']
```
